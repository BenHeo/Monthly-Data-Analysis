## 대여소, 반납소 조합에서 일정 수치에 못 미치는 조합은 삭제 (아마도 상당수 삭제 될 듯)

### 대여소와 반납소가 같은 경우는 이번 분석의 목적에 맞지 않으므로 제거 
(여기서 제거된 대여(반납)소의 경우 주변이 여가생활을 즐기기에 좋은 곳이라고 가정해볼 수 있겠다

```python
print(bike[bike['대여소번호']==bike['반납대여소번호']].shape)
bike[bike['대여소번호']==bike['반납대여소번호']].head()
```

```python
bike_notsame = bike[bike['대여소번호']!=bike['반납대여소번호']]
bike_notsame.head()
```

### 목적이 있는 분석이기 때문에 전체 기간 중 대여/반납 건수가 적은 대여소 미리 찾아 제거한다.

```python
bike_IO_counts = bike_notsame.groupby(['대여소번호', '반납대여소번호']).count()[['자전거번호']] # 임시로 자전거번호만
```

```python
print(bike_IO_counts.shape)
bike_IO_counts.head()
```

```python
sns.distplot(bike_IO_counts['자전거번호'], hist=False)
bike_IO_counts['자전거번호'].quantile([0.1*x for x in range(11)]) # 상위 20%부터 조합 이용 횟수가 14를 넘는다
```

```python
bike_IO_counts[bike_IO_counts['자전거번호']==8196]
```

```python
place_info.loc[(place_info['대여소번호']==183) | (place_info['대여소번호']==113)] # 코오롱아파트에 무슨 일이...
```

### 나름 긴 기간동안 조합 이용 횟수가 100회가 되지 않는다면 규칙적 패턴이 있다고 보기 어려울 수 있다

```python
bike_IO_over100 = bike_IO_counts[bike_IO_counts['자전거번호'] >= 100]
print(bike_IO_over100.shape)
bike_IO_over100.head()
```

```python
sns.distplot(bike_IO_over100['자전거번호'], hist=False)
bike_IO_over100['자전거번호'].quantile([0.1*x for x in range(11)]) 
```

```python
# 목표는 따릉이를 이용하는 사람들 다수의 목적을 아는 것이다
sns.distplot(bike_IO_over100['자전거번호'], hist=False).set(xlim=(0, 1000))
```

### 200 혹은 300 이상으로 끊는 게 좋아보인다

```python
bike_IO_over300 = bike_IO_over100[bike_IO_over100['자전거번호'] >= 300]
print(bike_IO_over300.shape)
bike_IO_over300.head()
```

```python
bike_over300 = bike_IO_over300.reset_index().rename(columns = {'자전거번호':'공유횟수'})
bike_over300.head()
```

```python
bike_over300[(bike_over300['대여소번호'] == 230) & (bike_over300['반납대여소번호'] == 232)]
```

```python
bike2= pd.merge(bike_over300, bike_notsame, how='inner', on = ['대여소번호', '반납대여소번호'])
```
